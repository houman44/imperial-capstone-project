```python
import os
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import layers, models
import numpy as np

# Create sequential model
model = Sequential()

"""
    Define a single Dense (fully connected) layer named L_1 with the following specifications:

    Number of output units: 2

    Input shape: vectors of length 2

    Include bias terms (use_bias=True)
"""

L_1 = Dense(
    units=2,              # Number of output units
    input_shape=(2,),     # Input vectors of length 2
    use_bias=True,        # Include bias terms
    name='L_1'           # Layer name
)

model.add(L_1)

# Display model summary
model.summary()

print(L_1)
```
